```python
import os
from anthropic import Anthropic
```

```python
client = Anthropic(
    api_key=os.getenv("ANTHROPIC_API_KEY")
)
```

```python
messages = []
messages.append({
    "role":"user",
    "content":"hello claude"
})
response = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=1024,
    messages = messages
)
print(response.content[0].text)
```
